3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Функция преобразования опыта работы в месяца
def get_experience(data):
    years_words = ['год', 'года', 'лет']
    months_words = ['месяц', 'месяца', 'месяцев']
    year = 0
    month = 0
    if data == 'Не указано' or data is np.nan:
        return np.nan
    exp_df = data.split(' ')[:7]
    for i, item in enumerate(exp_df):
        if item in years_words:
            year = int(exp_df[i - 1])
        elif item in months_words:
            month = int(exp_df[i - 1])
        else: 
            continue    
    return year*12 + month
# Создаем новый столбец с опытом работы в месяцах
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience)
# Выводим медианное значение столбца
display(hh_df['Опыт работы (месяц)'].median())
# Удаляяем столбец
hh_df =hh_df.drop('Опыт работы', axis = 1)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Создаем функцию преобразования городоа в группы
def town(data):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    town_df = str(data.split(' , ')[0])
    if town_df in million_cities:
        return 'Город милионник'
    elif town_df in ['Москва', 'Санкт-Петербург']:
        return town_df
    else:
        return 'Другие'
# Создаем новый столбец с городами 
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(town)
# Выводим долю людей их Санкт-Петербурга в общей массе 
# display(hh_df[hh_df['Город'] == 'Санкт-Петербург'].shape[0] / hh_df['Город'].shape[0] * 100)
# display(hh_df['Город'].shape[0])

# Создаем функцию преобразования данных о переезде
def migration(data):
    migration_df = str(data.split(' , ')[-2].split(' ')[0])
    if migration_df == 'не':
        return False
    else:
        return True
# Создаем столбей со значениями bool
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(migration)
# display(hh_df[hh_df['Переезд'] == True].shape[0])

# Создаем функцию преобразования данных о переезде
def trip(data):
    trip_df = str(data.split(' , '))
    if 'не' in trip_df:
        return False
    else:
        return True
# Создаем столбец со значениями bool
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(trip)
# display(hh_df[hh_df['Командировки'] == True].shape[0])

# Доля готовности к переезду и командировкам из общей массы
display(hh_df[(hh_df['Готовность к переезду'] == True) & (hh_df['Готовность к командировкам'] == True)].shape[0]/ (hh_df['Готовность к переезду'].shape[0] + hh_df['Готовность к командировкам'].shape[0]) * 100)
# Удаляем столбец исходного образца
hh_df = hh_df.drop('Город, переезд, командировки', axis = 1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Создаем списки из занятости и графика работы 
busyness = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedule = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
# Проверяем наличие элементов из списков в столбце
for i in busyness:
    hh_df[i] = hh_df['Занятость'].apply(lambda x: i in x)
for w in schedule:
    hh_df[w] = hh_df['График'].apply(lambda x: w in x)
# Выводим количество людей с определенной занятостью
display(hh_df[(hh_df['проектная работа'] == True) & (hh_df['волонтерство'] == True)].shape[0])     
# Выводим количество людей с определенным расписанием 
display(hh_df[hh_df['вахтовый метод'] & hh_df['гибкий график']].shape[0])

# Удаляем столбцы
hh_df = hh_df.drop(['Занятость', 'График'], axis = 1)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Читаем файл
rate_df = pd.read_csv('/Users/arturboldacev/SF_Data_Science/PROJECT_1. Анализ вакансий на HeadHunter/data/ExchangeRates.csv')

# Преобразуем столбцы к типу datetime
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме']).dt.date
rate_df['date'] = pd.to_datetime(rate_df['date']).dt.date

# Создаем функцию преобразования валюты
def rate(data):
    currency = str(data.split(' ')[-1]).replace('.', '')
    if currency == 'грн':
        return 'UAH'
    elif currency == 'белруб':
        return 'BYN'
    elif currency == 'сум':
        return 'UZS'
    elif currency == 'руб':
        return 'RUB'
    else:
        return currency
# Создаем новый столбец с валютой    
hh_df['Currency'] = hh_df['ЗП'].apply(rate)
# Создаем новый столбец с зарплатиой в рублях
hh_df['ЗП руб'] = hh_df['ЗП'].apply(lambda x: float(x.split(' ')[0]))

# Объединяем столбцы
merged_hh_df = hh_df.merge(rate_df,
                           left_on = ['Currency', 'Обновление резюме'],
                           right_on = ['currency', 'date'],
                           how = 'left')
# Заполняем пропорцию рубля 
merged_hh_df['proportion'] = merged_hh_df['proportion'].fillna(1)
# Заполняем закрытие рубля
merged_hh_df['close'] = merged_hh_df['close'].fillna(1)

# Конвертируем ожидаемую зарплату в рубли
hh_df['ЗП руб'] = merged_hh_df['ЗП руб'] * (merged_hh_df['close'] / merged_hh_df['proportion'])
# Выводим медианное значение столбца
display(hh_df['ЗП руб'].median())
```

```python
# Удаляем столбцы с исходными значениями зарплаты и валюты
hh_df = hh_df.drop('ЗП', axis = 1)
hh_df = hh_df.drop('Currency', axis = 1)
display(hh_df.info())
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Создаем функцию метода z-отклонений
def outliers_z_score(data, feature, log_scale=False, left = 3, right = 3):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
# Применяем метод к столбцу
outliers, cleaned = outliers_z_score(hh_df, 'Возраст', log_scale = True, right = 4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
# Удаляем строки с выбросами
hh_df = hh_df.drop(list(outliers.index))

# Строим гистограмму логорофмированного масштаба
log_age = np.log(hh_df['Возраст'] + 1)
histplot = sns.histplot(log_age)
histplot.axvline(log_age.mean(), color = 'k', lw = 2)
histplot.axvline(log_age.mean() - 3 * log_age.std(), color = 'k', lw = 2)
histplot.axvline(log_age.mean() + 3 * log_age.std(), color = 'k', lw = 2)
```

Логарифмическое распределение ассемитрично в левую сторону.
